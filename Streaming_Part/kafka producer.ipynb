```python
from kafka import KafkaProducer
import time
import csv
import json
from datetime import datetime  # Import datetime module

# Create a Kafka producer instance
producer = KafkaProducer(
    bootstrap_servers='localhost:9092',
    value_serializer=lambda v: json.dumps(v).encode('utf-8')
)

# Open and read the CSV file
with open('last10k.csv', newline='', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    
    for row in reader:
        # Get current date and time from the machine
        now = datetime.now()
        row['Date'] = now.strftime('%Y-%m-%d')  # Example: '2025-04-15'
        row['Time'] = now.strftime('%H:%M:%S')  # Example: '14:35:22'

        # Send each row to Kafka
        try:
            producer.send('skippy', value=row)
            print(f"Sent: {row}")
        except Exception as e:
            print(f"Error sending message: {e}")
        
        time.sleep(2)  # Simulate streaming delay

# Close the producer once finished
producer.flush()
producer.close()
```
